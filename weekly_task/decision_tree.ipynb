```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset_Decision_Tree.csv")
df.head()
pd.set_option('display.max_rows', None)
```

```python
inputs = df.drop('Purchased',axis='columns')
```

```python
target = df['Purchased']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Gender = LabelEncoder()
le_Umur = LabelEncoder()
le_EstimatedSalary = LabelEncoder()
```

```python

inputs['gender_n'] = le_Gender.fit_transform(inputs['Gender'])
inputs['umur_n'] = le_Umur.fit_transform(inputs['Umur'])
inputs['estimatedSalary_n'] = le_EstimatedSalary.fit_transform(inputs['EstimatedSalary'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['User ID','Gender','Umur','EstimatedSalary'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,3,4]])
```
